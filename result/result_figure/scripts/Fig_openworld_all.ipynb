```python
import numpy as np
def create_multi_bars(labels, datas, tick_step=1,legend=[], group_gap=0.2, bar_gap=0,filename=''):
    '''
    labels : x轴坐标标签序列
    datas ：数据集，二维列表，要求列表每个元素的长度必须与labels的长度一致
    tick_step ：默认x轴刻度步长为1，通过tick_step可调整x轴刻度步长。
    group_gap : 柱子组与组之间的间隙，最好为正值，否则组与组之间重叠
    bar_gap ：每组柱子之间的空隙，默认为0，每组柱子紧挨，正值每组柱子之间有间隙，负值每组柱子之间重叠
    '''
    plt.figure(figsize=(6, 6))
    # fig, ax = plt.subplots()
    # ticks为x轴刻度
    # plt.ylim(4, 4.2)

    patterns = [ "/" , "\\" , "|" , "-" , "+" , "x", "o", "O", ".", "*" ]
    ticks = np.arange(len(labels)) * tick_step
    # group_num为数据的组数，即每组柱子的柱子个数
    group_num = len(datas)
    # group_width为每组柱子的总宽度，group_gap 为柱子组与组之间的间隙。
    group_width = tick_step - group_gap
    # bar_span为每组柱子之间在x轴上的距离，即柱子宽度和间隙的总和
    bar_span = group_width / group_num
    # bar_width为每个柱子的实际宽度
    bar_width = bar_span - bar_gap
    # baseline_x为每组柱子第一个柱子的基准x轴位置，随后的柱子依次递增bar_span即可
    baseline_x = ticks - (group_width - bar_span) / 2
    for index, y in enumerate(datas):
        plt.bar(baseline_x + index*bar_span, y, bar_width,hatch=patterns[index])
    plt.ylabel('NT2T rate',fontdict=font)

    # plt.title('multi datasets')
    # x轴刻度标签位置与x轴刻度一致
    plt.xticks(ticks, labels,size = 24)
    # plt.xlabel('Open-Domain Chatbot Models',fontdict=font)  # x轴标题
    plt.yticks(size = 24)
    plt.legend(['Ehanced with Org','Default'],fontsize=20,prop={'size':20},loc='lower left')
    # plt.savefig(filename, bbox_inches='tight',  pad_inches = 0)
    # plt.hlines(4.05, -0.5, 4.5, linewidth = 4,color = 'grey',linestyles ='--')
    plt.savefig(os.path.join(figure_path,filename), dpi=300,bbox_inches='tight',  pad_inches = 0)
    
    plt.show()
new_combined_data=[
    numpy.add(all_data_list_e[2],all_data_list_e[4])*0.5,
    numpy.add(all_data_list_e[0],all_data_list_e[3])*0.5
]
create_multi_bars(['DialoGPT-L', 'BB-s', 'BB-L'],new_combined_data,legend=[],bar_gap=0.02,filename='Fig4_b.pdf')
```

```python
import numpy as np
def create_multi_bars(labels, datas, tick_step=1,legend=[], group_gap=0.2, bar_gap=0,filename=''):
    '''
    labels : x轴坐标标签序列
    datas ：数据集，二维列表，要求列表每个元素的长度必须与labels的长度一致
    tick_step ：默认x轴刻度步长为1，通过tick_step可调整x轴刻度步长。
    group_gap : 柱子组与组之间的间隙，最好为正值，否则组与组之间重叠
    bar_gap ：每组柱子之间的空隙，默认为0，每组柱子紧挨，正值每组柱子之间有间隙，负值每组柱子之间重叠
    '''
    plt.figure(figsize=(6, 6))
    # fig, ax = plt.subplots()
    # ticks为x轴刻度
    # plt.ylim(4, 4.2)

    ticks = np.arange(len(labels)) * tick_step
    # group_num为数据的组数，即每组柱子的柱子个数
    group_num = len(datas)
    # group_width为每组柱子的总宽度，group_gap 为柱子组与组之间的间隙。
    group_width = tick_step - group_gap
    # bar_span为每组柱子之间在x轴上的距离，即柱子宽度和间隙的总和
    bar_span = group_width / group_num
    # bar_width为每个柱子的实际宽度
    bar_width = bar_span - bar_gap
    # baseline_x为每组柱子第一个柱子的基准x轴位置，随后的柱子依次递增bar_span即可
    baseline_x = ticks - (group_width - bar_span) / 2
    for index, y in enumerate(datas):
        plt.bar(baseline_x + index*bar_span, y, bar_width)
    plt.ylabel('ASRate',fontdict=font)

    # plt.title('multi datasets')
    # x轴刻度标签位置与x轴刻度一致
    plt.xticks(ticks, labels,size = 24)
    # plt.xlabel('Open-World Chatbot Models',fontdict=font)  # x轴标题
    plt.yticks(size = 24)
    plt.legend(['Ehanced with Prefix','Default'],fontsize=20,prop={'size':20},loc='lower right')
    # plt.savefig(filename, bbox_inches='tight',  pad_inches = 0)
    # plt.hlines(4.05, -0.5, 4.5, linewidth = 4,color = 'grey',linestyles ='--')
    plt.savefig(os.path.join(figure_path,filename), dpi=300,bbox_inches='tight',  pad_inches = 0)
    
    plt.show()
new_combined_data=[
    all_data_list_e[0],
    numpy.add(all_data_list_e[1],all_data_list_e[2])*0.5
]
create_multi_bars(list(key),new_combined_data,legend=[],bar_gap=0.02,filename='Fig4_d.pdf')
```

```python
import numpy as np
def create_multi_bars(labels, datas, tick_step=1,legend=[], group_gap=0.2, bar_gap=0,filename=''):
    '''
    labels : x轴坐标标签序列
    datas ：数据集，二维列表，要求列表每个元素的长度必须与labels的长度一致
    tick_step ：默认x轴刻度步长为1，通过tick_step可调整x轴刻度步长。
    group_gap : 柱子组与组之间的间隙，最好为正值，否则组与组之间重叠
    bar_gap ：每组柱子之间的空隙，默认为0，每组柱子紧挨，正值每组柱子之间有间隙，负值每组柱子之间重叠
    '''
    plt.figure(figsize=(6, 6))
    # fig, ax = plt.subplots()
    # ticks为x轴刻度
    # plt.ylim(4, 4.2)

    patterns = [ "/" , "\\" , "|" , "-" , "+" , "x", "o", "O", ".", "*" ]
    ticks = np.arange(len(labels)) * tick_step
    # group_num为数据的组数，即每组柱子的柱子个数
    group_num = len(datas)
    # group_width为每组柱子的总宽度，group_gap 为柱子组与组之间的间隙。
    group_width = tick_step - group_gap
    # bar_span为每组柱子之间在x轴上的距离，即柱子宽度和间隙的总和
    bar_span = group_width / group_num
    # bar_width为每个柱子的实际宽度
    bar_width = bar_span - bar_gap
    # baseline_x为每组柱子第一个柱子的基准x轴位置，随后的柱子依次递增bar_span即可
    baseline_x = ticks - (group_width - bar_span) / 2
    for index, y in enumerate(datas):
        plt.bar(baseline_x + index*bar_span, y, bar_width,hatch=patterns[index])
    plt.ylabel('NT2T rate',fontdict=font)

    # plt.title('multi datasets')
    # x轴刻度标签位置与x轴刻度一致
    plt.xticks(ticks, labels,size = 24)
    # plt.xlabel('Open-Domain Chatbot Models',fontdict=font)  # x轴标题
    plt.yticks(size = 24)
    plt.legend(['Ehanced with Prompt','Default'],fontsize=20,prop={'size':20},loc='lower left')
    # plt.savefig(filename, bbox_inches='tight',  pad_inches = 0)
    # plt.hlines(4.05, -0.5, 4.5, linewidth = 4,color = 'grey',linestyles ='--')
    plt.savefig(os.path.join(figure_path,filename), dpi=300,bbox_inches='tight',  pad_inches = 0)
    
    plt.show()
new_combined_data=[
    all_data_list_e[0],
    numpy.add(all_data_list_e[1],all_data_list_e[2])*0.5
]
create_multi_bars(['DialoGPT-L', 'BB-s', 'BB-L'],new_combined_data,legend=[],bar_gap=0.02,filename='Fig4_d.pdf')
```
